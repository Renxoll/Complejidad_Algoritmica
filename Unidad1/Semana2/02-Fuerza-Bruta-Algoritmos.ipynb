### Selection Sort

```python
def selectionSort(L:list):
    n = len(L)
    for i in range(n-1):
        min_pos = i
        for j in range(i+1, n):
            if L[j] < L[min_pos]:
                min_pos = j
                
        if min_pos != i:
            L[min_pos], L[i] = L[i], L[min_pos]
```

#### Bubble Sort

```python
def bubbleSort(L2):
    n = len(L2)
    for i in range (n-1):
        for j in range (n-1-i):
            if L2[j] > L2[j+1]:
                L2[j], L2[j+1] = L2[j+1], L2[j]
            
```

```python
D = [r.randint(1,100) for _ in range(10) ]
print(D)
```

```python
bubbleSort(D)
print(D)
```

```python
from itertools import permutations

def is_valid_permutation(p):
    n = len(p) // 2
    for i in range(1, n + 1):
        positions = [idx for idx, val in enumerate(p) if val == i]
        if len(positions) != 2 or (positions[1] - positions[0]) != i:
            return False
    return True

def generate_combinations(n):
    elements = [i for i in range(1, n + 1) for _ in range(2)]
    all_permutations = set(permutations(elements))
    valid_permutations = [p for p in all_permutations if is_valid_permutation(p)]
    return valid_permutations

def main():
    n = int(input("Ingrese el valor de n (un número positivo): "))
    if n <= 0:
        print("El valor de n debe ser un número positivo.")
        return

    combinations = generate_combinations(n)
    if combinations:
        print(f"Se encontraron {len(combinations)} combinaciones válidas:")
        for combination in combinations:
            print(combination)
    else:
        print("No se encontraron combinaciones válidas.")

if __name__ == "__main__":
    main()
```
